```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la seal senoidal
frecuencia_angular = 2 * np.pi  # Frecuencia angular (0)
tiempo = np.linspace(0, 2 * np.pi, 1000)  # Vector de tiempo

# Funcin seno
seno = np.sin(frecuencia_angular * tiempo)

# Transformada de Fourier
transformada = np.fft.fft(seno)
frecuencias = np.fft.fftfreq(len(tiempo), d=(tiempo[1] - tiempo[0]))

# Grfico de la seal y su transformada
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(tiempo, seno, label='Seal seno')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Seal seno en el dominio del tiempo')

plt.subplot(2, 1, 2)
plt.plot(frecuencias, transformada, label='Transformada de Fourier')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.title('Transformada de Fourier de la seal seno')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# Cargamos datos
df = pd.read_csv('/content/Market2020.csv')

# Columna Gold es la que usaremos
data = df['Gold'].values

# Agregamos ruido gaussiano
mu, sigma = 0, np.std(data) * 0.05
noise = np.random.normal(mu, sigma, data.shape)
noisy_signal = data + noise

# Aplicamos un filtro de medias mviles
window_size = 10
smooth_signal = lfilter(np.ones(window_size) / window_size, 1, noisy_signal)

# Graficamos la seal original, la seal ruidosa y la seal suavizada
plt.figure(figsize=(10, 6))
plt.plot(data, label='Seal original', color='blue')
plt.plot(noisy_signal, label='Seal ruidosa', color='orange')
plt.plot(smooth_signal, label='Seal suavizada', color='green')
plt.title('Seal original vs. Seal ruidosa vs. Seal suavizada')
plt.xlabel('ndice de tiempo')
plt.ylabel('Valor')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# Generar la secuencia X
i = np.arange(1, 5001)
X = np.cos(0.001*i) + 0.5*np.cos(0.01*i - 1)

# Calcular la Transformada de Fourier
TF = fft(X)

# Obtener los valores absolutos
abs_TF = np.abs(TF)

# Normalizar los valores absolutos
abs_TF = abs_TF / np.max(abs_TF)

# Graficar los valores absolutos de la Transformada de Fourier
plt.figure(figsize=(10, 5))
plt.plot(abs_TF[:300])
plt.title('Valores Absolutos de la Transformada de Fourier del Coseno')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# Definir el rango de i
i = np.arange(1, 5001)

# Definir las funciones
x1 = np.cos(0.001 * i)
x2 = 0.5 * np.cos(0.01 * i - 1)

# Calcular el vector X
X = x1 + x2

# Agregamos ruido gaussiano
mu, sigma = 0, np.std(X) * 0.05
noise = np.random.normal(mu, sigma, X.shape)
noisy_signal = X + noise

# Aplicamos un filtro de medias mviles
window_size = 10
smooth_signal = lfilter(np.ones(window_size) / window_size, 1, noisy_signal)

# Graficamos la seal original, la seal ruidosa y la seal suavizada
plt.figure(figsize=(10, 6))
plt.plot(X, label='Seal original', color='blue')
plt.plot(noisy_signal, label='Seal ruidosa', color='orange')
plt.plot(smooth_signal, label='Seal suavizada', color='green')
plt.title('Seal original vs. Seal ruidosa vs. Seal suavizada')
plt.xlabel('ndice de tiempo')
plt.ylabel('Valor')
plt.grid(True)
plt.legend()
plt.show()
```
